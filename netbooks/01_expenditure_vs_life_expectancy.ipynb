# Hypothesis 1: Health Expenditure vs Life Expectancy

**Hypothesis:**  
Countries that spend more on health tend to have higher life expectancy,  
but the marginal benefit decreases at higher spending levels.

**Goal:**  
Analyze the relationship between health expenditure and life expectancy,
controlling for economic size.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/processed/final/final_enriched_dataset.csv")
```

```python
df[[
    "life_expectancy",
    "health_expenditure_pct_gdp",
    "log_gdp_per_capita"
]].describe()
```

```python
df[[
    "life_expectancy",
    "health_expenditure_pct_gdp",
    "log_gdp_per_capita"
]].corr()
```

```python
sns.scatterplot(
    data=df,
    x="health_expenditure_pct_gdp",
    y="life_expectancy",
    alpha=0.3
)
plt.title("Health Expenditure vs Life Expectancy")
plt.show()
```

```python
sns.scatterplot(
    data=df,
    x="health_expenditure_pct_gdp",
    y="life_expectancy",
    hue="log_gdp_per_capita",
    alpha=0.4,
    palette="viridis"
)
plt.title("Health Expenditure vs Life Expectancy (Controlled by GDP)")
plt.show()
```

### Conclusion

- The relationship between health expenditure and life expectancy is positive.
- Diminishing returns are visible at higher spending levels.
- Economic context explains part, but not all, of the variation.

**Hypothesis supported**, with clear evidence of non-linearity.
