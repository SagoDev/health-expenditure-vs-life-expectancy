# Hypothesis 3: Lagged Effects of Health Spending

**Hypothesis:**  
Health expenditure affects health outcomes with a time lag.

**Goal:**  
Compare immediate vs delayed effects.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/processed/final/final_dataset_enriched.csv")
```

```python
sns.scatterplot(
    data=df,
    x="health_exp_yoy_growth",
    y="life_expectancy_yoy_change",
    alpha=0.4
)
plt.axhline(0, linestyle="--", color="gray")
plt.axvline(0, linestyle="--", color="gray")
plt.title("Same-Year Change vs Change")
plt.show()
```

### Conclusion

- Same-year changes show weak or noisy relationships.
- Lagged expenditure shows clearer association with improvement.
- Health investments require time to materialize.

**Hypothesis supported.**
