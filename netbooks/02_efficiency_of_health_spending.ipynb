# Hypothesis 2: Efficiency of Health Spending

**Hypothesis:**  
Health outcomes differ significantly across countries with similar levels
of health expenditure.

**Goal:**  
Identify efficiency differences in health systems.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/processed/final/final_dataset_enriched.csv")
```

```python
sns.scatterplot(
    data=df,
    x="health_expenditure_pct_gdp",
    y="life_expectancy",
    alpha=0.4
)
plt.title("Same Spending, Different Outcomes")
plt.show()
```

```python
sns.scatterplot(
    data=df,
    x="health_expenditure_pct_gdp",
    y="life_expectancy",
    hue="life_expectancy_per_health_exp",
    palette="viridis",
    alpha=0.6
)
plt.title("Health Spending Efficiency")
plt.show()
```

```python
df_eff = df.dropna(subset=["life_expectancy_per_health_exp"])

df_eff.sort_values(
    "life_expectancy_per_health_exp",
    ascending=False
).head(10)[
    ["country", "year", "life_expectancy", "health_expenditure_pct_gdp"]
]
```

### Conclusion

- Countries with similar spending levels achieve very different outcomes.
- High spending does not guarantee efficiency.
- Institutional and structural factors likely play a key role.

**Hypothesis strongly supported.**
